# 4.1 - Bài 1: y=ax+b

```python
def tinh_y(a, b, x):
    lst = []

    for i in range(0,len(x)):
        ele = a*i + b
        lst.append(ele) 
    
    return lst
```

```python
def ve_PT():
    a = int(input("Nhập hệ số a: "))
    b = int(input("Nhập hệ số b: "))
    x = list(range(0,5,1))
    y = tinh_y(a,b,x)
    
    plt.plot(x,y,marker="o")
    plt.title("Đồ thị hàm số y=ax+b", fontsize=14, fontweight="bold", color="red")
    plt.xlabel("x", fontsize=12, color="red")
    plt.ylabel("y", fontsize=12, color="red")
    plt.show()
```

```python
ve_PT()
```

# 4.1 - Bài 2: Đọc và vẽ biểu đồ chứng khoán

```python
import csv

lst_date_csv = []
lst_openFixed_csv = []
lst_closeFixed_csv = []

with open("data/excel_flc.csv", mode='r', encoding='utf-8') as f:
    reader = csv.reader(f)
    next(reader)
    
    for row in reader:
        lst_date_csv.append(row[1])
        lst_openFixed_csv.append(row[2])
        lst_closeFixed_csv.append(row[5])
```

```python
import datetime as dt

lst_date_format = [dt.datetime.strptime(item, '%Y%m%d').date() for item in lst_date_csv]
```

```python
import matplotlib
import matplotlib.pyplot as plt

plt.plot(lst_date_format[:25], sorted(lst_openFixed_csv[:25]), color="blue", marker="o")
plt.plot(lst_date_format[:25], sorted(lst_closeFixed_csv[:25]), color="red", marker="o")
plt.title("Biểu đồ thể hiện giá mở cửa và giá đóng cửa theo ngày")
plt.xlabel("Ngày")
plt.ylabel("Giá tiền")
plt.show()
```
