```python
import matplotlib
import matplotlib.pyplot as plt
import csv
import datetime as dt
from math import sqrt, pow
import numpy as np
```

# 4.1 - Bài 1: y=ax+b

```python
def ve_PT():
    a = int(input("Nhập hệ số a: "))
    b = int(input("Nhập hệ số b: "))
    x_values = list(range(-1,4,1))
    y_values = [(a*x + b) for x in x_values]
    
    plt.plot(x_values, y_values, marker="o")
    plt.title("Đồ thị hàm số y = ax+b", fontsize=14, fontweight="bold", color="red")
    plt.xlabel("x", fontsize=12, color="red")
    plt.ylabel("y", fontsize=12, color="red")
    plt.show()
```

```python
ve_PT()
```

# 4.1 - Bài 2: Đọc và vẽ biểu đồ chứng khoán

```python
def ve_BieuDoCK(file_dir,maCK):
    lst_date_csv = []
    lst_openFixed_csv = []
    lst_closeFixed_csv = []
    #lst_high_csv=[]
    #lst_low_csv=[]

    with open(file_dir, mode='r', encoding='utf-8') as f:
        reader = csv.reader(f)
        next(reader)

        for row in reader:
            lst_date_csv.append(row[1])
            lst_openFixed_csv.append(row[2])
            lst_closeFixed_csv.append(row[5])
            #lst_high_csv.append(row[3])
            #lst_low_csv.append(row[4])

    lst_date_format = [dt.datetime.strptime(item, '%Y%m%d').date() for item in lst_date_csv]

    lst_openFixed_csv =[float(x) for x in lst_openFixed_csv]
    lst_closeFixed_csv =[float(x) for x in lst_closeFixed_csv]
    
    fig = plt.figure(figsize=(12,8))
    ax1 = fig.add_subplot(2, 2, 1)
    ax1.plot(lst_date_format, lst_openFixed_csv, color="b")
    ax1.title.set_text('Biểu đồ thể hiện giá mở cửa qua từng năm')
    ax1.set_xlabel("Năm")

    ax2 = fig.add_subplot(2, 2, 2)
    ax2.plot(lst_date_format, lst_closeFixed_csv, color="r")
    ax2.title.set_text('Biểu đồ thể hiện giá đóng cửa qua từng năm')
    ax2.set_xlabel("Năm")
    
    fig.suptitle(maCK)
    plt.tight_layout()
    plt.show()
```

```python
ve_BieuDoCK("data/excel_flc.csv","FLC")
```

# 4.1 - Bài 3b: y=ax^2 + bx + c (-5 <= x <=5)

```python
a = int(input("Nhập hệ số a: "))
b = int(input("Nhập hệ số b: "))
c = int(input("Nhập hệ số c: "))

x_values = np.linspace(-5, 5, 100)
y_values = [(a*pow(x,2) + b*x + c) for x in x_values]

ax = plt.axes()
ax.plot(x_values, y_values)

plt.title("Đồ thị hàm số y = ax^2 + bx + c", fontsize=14, fontweight="bold", color="r")
plt.xlabel("Giá trị x", fontsize=12, color="b")
plt.ylabel("Giá trị y", fontsize=12, color="b")

plt.show()
```
