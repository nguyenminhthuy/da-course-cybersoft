```python
import matplotlib
import matplotlib.pyplot as plt
import csv
import datetime as dt
from math import sqrt, pow
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/black-friday-sales-eda.csv')
df.values
```

```python
counts_gender_byage = df.groupby(['Age', 'Gender']).Age.count().unstack()
print(counts_gender_byage)
```

```python
ax = counts_gender_byage.plot(kind='bar',stacked = False, colormap = 'Paired')

for p in ax.patches:
        ax.annotate(np.round(p.get_height(),decimals=0).astype(np.int64), 
                    (p.get_x()+p.get_width()/2., p.get_height()), ha='center', 
                    va='center', xytext=(2, 10), textcoords='offset points')

plt.xlabel ('Nhóm tuổi')
plt.ylabel ('Số lượng mua hàng')
plt.title('Số lượng mua hàng của từng nhóm tuổi theo giới tính',fontsize=14)
plt.show()

print('Males of age between 18-45 are purchasing more items.')
```
