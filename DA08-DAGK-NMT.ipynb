```python
import matplotlib
import matplotlib.pyplot as plt
import csv
import datetime as dt
from math import sqrt, pow
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('data/black-friday-sales-eda.csv')
```

```python
counts_gender_byage = df.groupby(['Age', 'Gender']).Age.count().unstack()
print(counts_gender_byage)
```

```python
counts_city_byage = df.groupby(['City_Category', 'Age']).Age.count().unstack()
print(counts_city_byage)
```

```python
counts_city_bygendermst = df.groupby(['City_Category', 'Gender','Marital_Status']).City_Category.count().unstack()
print(counts_city_bygendermst)
```

```python
counts_product1_bygender = df.groupby(['Product_Category_1', 'Gender']).Product_Category_1.count().unstack()
print(counts_product1_bygender)
```

```python
counts_product2_bygender = df.groupby(['Product_Category_2', 'Gender']).Product_Category_2.count().unstack()
print(counts_product2_bygender)
```

```python
counts_product3_bygender = df.groupby(['Product_Category_3', 'Gender']).Product_Category_3.count().unstack()
print(counts_product3_bygender)
```

```python
ax = counts_gender_byage.plot(kind='bar',stacked = False, colormap = 'Paired')

plt.xlabel ('Nhóm tuổi')
plt.ylabel ('Số lượng')
plt.title('Số lượng mua hàng của Nam/Nữ theo nhóm tuổi',fontsize=14)
plt.show()

print('Males of age between 18-45 are purchasing more items.')
```

```python
ax = counts_city_byage.plot(kind='bar',stacked = False, colormap = 'Paired')

plt.xlabel ('Thành phố')
plt.ylabel ('Số lượng')
plt.title('Số lượng mua hàng của từng thành phố theo nhóm tuổi',fontsize=14)
plt.show()

print('People form City B are purchasing more items.')
```

```python
ax = counts_city_bygendermst.plot(kind='bar',stacked = False, colormap = 'Paired')

plt.xlabel ('(Thành phố, Giới tính)')
plt.ylabel ('Số lượng')
plt.title('Số lượng mua hàng của từng thành phố và giới tính theo tình trạng hôn nhân',fontsize=14)
plt.show()

print('')
```

```python
ax = counts_product1_bygender.plot(kind='bar',stacked = False, colormap = 'Paired')

plt.xlabel ('Product Category 1')
plt.ylabel ('Số lượng')
plt.title('Số lượng mua hàng của danh mục sản phẩm 1 theo giới tính',fontsize=14)
plt.show()

print('')
```

```python
ax = counts_product2_bygender.plot(kind='bar',stacked = False, colormap = 'Paired')

plt.xlabel ('Product Category 2')
plt.ylabel ('Số lượng')
plt.title('Số lượng mua hàng của danh mục sản phẩm 2 theo giới tính',fontsize=14)
plt.show()

print('')
```
