```python
import matplotlib.pyplot as plt
```

```python
# Vẽ phương trình y = ax + b
def plot_linear(a,b):
    x = [-10,10]
    y = [a*_x + b for _x in x ]
    
    fig = plt.figure(figsize = (5,5))
    plt.plot(x, y, color="navy", lw=3)
    
    plt.grid(ls="--")
    
    plt.axis("equal")
    plt.title(f"Biểu đồ phương trình y={a}x {b:+}")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```

```python
plot_linear(0.5,-10)
```

```python
def plot_parabol(a,b,c):
    x = range(-10,10,1)
    y = [a*_x*_x + b*_x + c for _x in x]    
    
    fig = plt.figure(figsize=(5,5))
    plt.plot(x,y)
    plt.axis('equal')
    plt.grid(ls="--")
    plt.title(f"Biểu đồ phương trình y={a}x^2 {b:+}x {c:+}")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()
```

```python
plot_parabol(0.5,0.9,-15)
```

```python
import csv
```

```python
with open("data-stock/excel_acb.csv",mode='r', encoding="utf-8", newline='') as f:
    data = csv.reader(f)
    header = next(data)
    i_close = header.index("<Close>")
    close = []    
    for row in data:
        close.append(float(row[i_close]))
        
close = close[::-1]
```

```python
def stock_plot(brand,close):
    plt.plot(close, color="teal")
    plt.title(f"Stock price of {brand}")
    plt.ylabel("price")
    plt.xlabel("date")
    plt.tight_layout()
```

```python
stock_plot("ACB",close)
```

```python
list_brand = ["acb","dxg","flc","fpt","vjc","vnm"]
```

```python
list_price = []
for brand in list_brand:
    with open(f"data-stock/excel_{brand}.csv",mode='r', encoding="utf-8", newline='') as f:
        data = csv.reader(f)
        header = next(data)
        i_close = header.index("<Close>")
        close = []    
        for row in data:
            close.append(float(row[i_close]))
        close = close[::-1]
            
    list_price.append(close)
```

```python
fig = plt.figure(figsize=(9,5))
for i,brand in enumerate(list_brand):
    ax = fig.add_subplot(2,3,i+1)
    stock_plot(str.upper(brand),list_price[i])
fig.suptitle("ABC XYZ Figure Title ")
plt.tight_layout()
plt.show()
```
