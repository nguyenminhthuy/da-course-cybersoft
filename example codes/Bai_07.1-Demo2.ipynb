```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
np.set_printoptions(precision=3, suppress=True)

print(pd.__version__)
```

# 1. Ghi/Đọc csv

```python
a = np.random.randint(0, 10, size=(10,4))
df1 = pd.DataFrame(data=a, index=range(1,11), columns=list('ABCD'), dtype='uint8')
df1
```

## 1.1. Ghi/Đọc CSV cơ bản

**Lưu vào file csv**

```python
df1.to_csv('data/df1.csv', index=False)
```

**Đọc từ file csv**

```python
df1_read = pd.read_csv('data/df1.csv')
df1_read
```

## 1.2. Ghi/Đọc với cột index

```python
df1.to_csv('data/df2.csv', index=True)
```

```python
# Set tham số chọn 1 cột làm index
df2_read = pd.read_csv('data/df2.csv', index_col=0)
df2_read
```

```python
# Chọn các cột sẽ dùng bằng thứ tự của cột
df3_read = pd.read_csv('data/df2.csv', index_col=0, usecols=[0,1,3])
df3_read
```

## 1.3. Append một file csv

```python
# (1) Tạo một file df3.csv
df4 = df1 * 2
df4.to_csv('data/df3.csv', mode='w', index=False)
```

```python
# (2) Lưu mảng df1 vào cùng file df3.csv
df1.to_csv('data/df3.csv', mode='a', index=False, header=False)
```

**Cách khác: tự động xác định nếu file chưa có dữ liệu thì thêm header**

```python
# Lưu ý: Trên Windows có thể bị thêm dòng trống, khi đó cần set newline là rỗng
with open('data/df3.csv', mode='a', newline='') as f:
    df1.to_csv(f, index=False, header=(f.tell()==0))
```

```python
df3_read = pd.read_csv('data/df3.csv')
df3_read
```

# 2. Ghi/Đọc Excel

**Lưu và đọc dataframe vào/từ file excel**

```python
df1.to_excel('data/df1_excel.xlsx', sheet_name='DF1', index=False)
```

```python
df1_read = pd.read_excel('data/df1_excel.xlsx', sheet_name='DF1', index_col=None)
```

```python
df1_read
```

**Mở rộng file đã có**

```python
# Mở rộng sheet mới
with pd.ExcelWriter('data/df1_excel.xlsx', engine='openpyxl', mode='a') as writer:
    df3_read.to_excel(writer, sheet_name='DF3')
```

```python
# Mở rộng sheet cũ? ==> Báo lỗi
with pd.ExcelWriter('data/df1_excel.xlsx', engine='openpyxl', mode='a') as writer:
    df3_read.to_excel(writer, sheet_name='DF1')
```

```python
# Có hướng dẫn dùng mode='A', nhưng không hoạt động!
with pd.ExcelWriter('data/df1_excel.xlsx', mode='A') as writer:
    df1.to_excel(writer, sheet_name='DF4')
```

```python
# Ghi vào nhiều sheets cùng lúc
with pd.ExcelWriter('data/df2_3_excel.xlsx') as writer:
    df4.to_excel(writer, sheet_name='DF2')
    df3_read.to_excel(writer, sheet_name='DF3')
```
