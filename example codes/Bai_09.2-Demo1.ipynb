# Phần 2. `Các loại biểu đồ trong Seaborn`

## Load bộ dữ liệu về 'tips' có sẵn của seaborn

```python
sns.get_dataset_names()
```

```python
tip_df = sns.load_dataset('tips')
tip_df
```

```python
tip_df.columns = ['Tổng đơn', 'Tip', 'Giới tính', 
                  'Hút thuốc', 'Ngày', 'Thời điểm', 'Số lượng']
tip_df
```

# 1. Relational plots (biểu đồ quan hệ)

## 1.1. Các hàm axes-level

**Scatterplot**

```python
# Cách truyền tham số thông thường
sns.scatterplot(
                x=tip_df['Tổng đơn'], y=tip_df['Tip'], 
               )
plt.show()
```

```python
# Truyền tham số bằng DataFrame
sns.scatterplot(data=tip_df, 
                x='Tổng đơn', y='Tip', 
               )
plt.show()
```

```python
# Tự động phân chia loại theo một cột nào đó (VD: 'Giới tính')
sns.scatterplot(data=tip_df, 
                x='Tổng đơn', y='Tip', 
                hue='Giới tính'
               )
plt.show()
```

```python
# Thay đổi kích thước và thứ tự hue
# hue_order = ['Female', 'Male']
fig = plt.figure(figsize=(7,5))
sns.scatterplot(data=tip_df, 
                x='Tổng đơn', y='Tip', 
                hue='Giới tính', hue_order=['Female', 'Male']
               )
plt.show()
```

**Lineplot**

```python
fig = plt.figure(figsize=(7,5))
sns.lineplot(data=tip_df, x='Tổng đơn', y='Tip', 
             hue='Giới tính', 
             palette=['r', 'g']
            )
plt.show()
```

## 1.2. Hàm ở Figure-level

**Hàm relplot()**

```python
# kind = 'line'/'scatter'
sns.relplot(data=tip_df, 
            x='Tổng đơn', y='Tip', 
            hue='Giới tính',
            kind='scatter', 
           )
plt.show()
```

**Thay đổi kích thước Figure**

```python
# fig = plt.figure(figsize=(107,5)) # Không có tác dụng
sns.relplot(data=tip_df, 
            x='Tổng đơn', y='Tip', 
            hue='Giới tính',
            
            # Muốn thay đổi size, phải set height và aspect
            height=5, aspect=7/5, 
           )
plt.show()
```

# 2. Distributional plots (các biểu đồ về phân bố dữ liệu)

### 2.1. Figure-Level

**Hàm `displot()`**

```python
sns.displot(data=tip_df, x='Tổng đơn', kde=False, bins=20)
plt.show()
```

```python
# Thêm đường biểu diễn ước lượng mật độ (kde)
# kde: kernel density estimate ("ước lượng mật độ hạt nhân/lõi")
sns.displot(data=tip_df, x='Tổng đơn', kde=True, bins=20)
plt.show()
```

### 2.2. Các hàm axes-level

**kdeplot (kernel density estimate)**

```python
# Vẽ biểu đồ kde với hàm kdeplot (axes-level)
sns.kdeplot(data=tip_df, x='Tổng đơn', hue='Giới tính', shade=True)
```

**histplot**

```python
# Vẽ biểu đồ histogram thông thường với hàm histplot (axes-level)
fig = plt.figure(figsize=(8,6))
sns.histplot(tip_df, x='Tip', kde=True, bins=20, hue='Giới tính')
```

**Hàm `countplot`**

```python
sns.countplot(x=tip_df["Giới tính"], hue=tip_df["Hút thuốc"], dodge=True)
plt.show()
```

**Biểu đồ dạng cummulative (tổng lũy kế/cộng dồn)**

Hàm **ecdfplot** (empirical cumulative distribution functions)

```python
sns.ecdfplot(data=tip_df, x='Số lượng', 
             stat='proportion' # proportion/count
            )
```

**Biểu đồ dạng phân bố dày (rugplot)**

```python
sns.rugplot(data=tip_df, x='Tổng đơn', y='Tip', hue='Giới tính')
plt.show()
```

```python
# Kết hợp 'rugplot' với các biểu đồ khác (Cùng là axes-level)
sns.regplot(data=tip_df, x='Tổng đơn', y='Tip')
sns.scatterplot(data=tip_df, x='Tổng đơn', y='Tip', hue='Giới tính')
sns.rugplot(data=tip_df, x='Tổng đơn', y='Tip', hue='Giới tính')
plt.show()
```

# 3. Categorical plots (Biểu đồ có tính phân loại)

### 3.1. Biểu đồ cột

**Yêu cầu: Vẽ biểu đồ trung bình tổng đơn theo ngày trong tuần, và phân biệt khách hàng theo giới tính**

```python
# Một số estimator: np.mean, np.median, np.sum
bieudo = sns.barplot(data=tip_df, 
                     x='Ngày', y='Tổng đơn', 
                     hue='Giới tính', 
                     estimator=np.mean)
plt.show()
```

```python
# Thay đổi error-bar
sns.barplot(data=tip_df, x='Ngày', y='Tổng đơn',
            hue='Giới tính',
            estimator=np.mean,
            ci=90, #None, khoảng dao động
            errcolor='r', capsize=0.2,
           )
plt.show()
```

```python
# Nếu không cần error_bar thì set ci=None
bieudo = sns.barplot(data=tip_df, x='Ngày', y='Tổng đơn', hue='Giới tính', 
                     errcolor='r', 
                     ci=None,
                     estimator=np.sum)
plt.show()
```

### 3.2. Biểu đồ dạng box - violin - point

**Boxplot**

```python
sns.boxplot(data=tip_df, x='Ngày', y='Tổng đơn', hue='Giới tính')
```

```python
# 'enhanced box' cho các dữ liệu lớn
# Tham khảo: https://seaborn.pydata.org/generated/seaborn.boxenplot.html
sns.boxenplot(data=tip_df, x='Ngày', y='Tổng đơn')
```

**Violinplot**

```python
sns.violinplot(data=tip_df, x='Ngày', y='Tổng đơn', hue='Giới tính')
```

**Pointplot**

```python
sns.pointplot(data=tip_df, x='Ngày', y='Tổng đơn')
```

### 3.3. Biểu đồ dạng cột - scatter

```python
# Biểu đồ dạng string: mảnh/dải (axes-level)
sns.stripplot(data=tip_df, x='Ngày', y='Tổng đơn', hue='Giới tính')
```

```python
# Biểu đồ dạng beeswarm: tổ ong (axes-level)
sns.swarmplot(data=tip_df, x='Ngày', y='Tổng đơn', hue='Giới tính')
```

```python
# Figure-Level
sns.catplot(data=tip_df, 
            x='Ngày', y='Tip', 
            hue='Giới tính')
plt.show()
```

```python
# Figure-level
sns.catplot(data=tip_df, x='Ngày', y='Tổng đơn', 
            height=4, aspect=6/4, 
            hue='Giới tính',
            jitter=False)
```

# 4. Regression plot (Biểu đồ hồi quy)

## 4.1. Axes level

```python
# Dùng hàm regplot()
sns.regplot(data=tip_df, x='Tổng đơn', y='Tip', color='g',
           # hue='Thời điểm'
           )
```

```python
# Thay đổi vùng của khoảng tin cậy CI
sns.regplot(data=tip_df, x='Tổng đơn', y='Tip', color='g',
            ci=90,
           # hue='Thời điểm'
           )
```

## 4.2. Figure level

```python
# Dùng hàm lmplot()
sns.lmplot(data=tip_df, x='Tổng đơn', y='Tip', 
           height=6, aspect=7/6,
           hue='Giới tính')
```

<h1><font color='red'>** Vấn đề cần lưu ý **</font></h1>

<h3> Các hàm figure-level không hỗ trợ ax, nên không vẽ chung trên cùng biểu đồ theo cách thông thường được </h3>

```python
fig, ax = plt.subplots()
sns.displot(tip_df, x="Tổng đơn", ax=ax)
sns.displot(tip_df, x="Tip", ax=ax)
```

**Chỉ có thể vẽ trên các axes một cách chủ động với các hàm axes-level**

```python
fig, ax = plt.subplots()
sns.histplot(tip_df, x="Tổng đơn", ax=ax, color='r')
sns.histplot(tip_df, x="Tip", ax=ax, color='g')

plt.show()
```
